```python
from vedo import settings
settings.default_backend = '2d'  # Para visualización inline

import numpy as np
import trimesh
import open3d as o3d
from vedo import Plotter, Sphere, color_map, Mesh, show, write
import numpy as np
import trimesh
import open3d as o3d
from vedo import Plotter, Sphere, color_map, Mesh, write

# Configurar semilla para reproducibilidad
np.random.seed(42)

# 1. Crear lista de coordenadas 3D
num_puntos = 30
puntos = np.random.uniform(-8, 8, size=(num_puntos, 3))

# 2. Generar primitivas 3D con variaciones
meshes = []

for i, pos in enumerate(puntos):
    # Variar color basado en la posición en Z
    color = color_map(pos[2], 'rainbow', vmin=-8, vmax=8)

    if i % 3 == 0:  # Cubos usando trimesh
        tamaño = 0.5 + abs(pos[0]) * 0.1
        cubo = trimesh.creation.box(extents=[tamaño]*3)
        cubo.apply_translation(pos)
        mesh = Mesh([cubo.vertices, cubo.faces]).color(color)

    elif i % 3 == 1:  # Esferas usando vedo
        radio = 0.3 + abs(pos[1]) * 0.1
        mesh = Sphere(pos=pos, r=radio).color(color)

    else:  # Cilindros usando open3d
        altura = 1.0 + abs(pos[2]) * 0.2
        cilindro = o3d.geometry.TriangleMesh.create_cylinder(
            radius=0.2, height=altura, resolution=20)
        cilindro.translate(pos)
        mesh = Mesh([np.asarray(cilindro.vertices),
                    np.asarray(cilindro.triangles)]).color(color)

    if pos[0] > 0:
        mesh.rotate(np.random.random()*360, axis=(0,1,0))

    meshes.append(mesh)

# Mostrar usando la función show() global de vedo
show(meshes,
     axes=14,
     viewup='z',
     title="Primitivas 3D Variadas",
     bg="white",
     bg2="lightblue",
     interactive=True,
     new=True)  # Fuerza nueva ventana

```

```python
import numpy as np
import trimesh
import open3d as o3d
from vedo import Plotter, Sphere, color_map, Mesh, write, merge, show
import os

# Configuración inicial
np.random.seed(42)
num_puntos = 30
puntos = np.random.uniform(-8, 8, size=(num_puntos, 3))
export_dir = "exportados/"
os.makedirs(export_dir, exist_ok=True)

# Lista para almacenar todos los objetos vedo
vedo_objects = []

for i, pos in enumerate(puntos):
    color = color_map(pos[2], 'rainbow', vmin=-8, vmax=8)

    # Cubos con trimesh (exportación individual)
    if i % 3 == 0:
        tamaño = 0.5 + abs(pos[0]) * 0.1
        cubo = trimesh.creation.box(extents=[tamaño]*3)
        cubo.apply_translation(pos)

        # Exportar con trimesh
        cubo.export(f"{export_dir}cubo_{i}.gltf")

        # Convertir a vedo y aplicar color
        vedo_mesh = Mesh([cubo.vertices, cubo.faces]).color(color)

    # Esferas con vedo (exportación individual)
    elif i % 3 == 1:
        radio = 0.3 + abs(pos[1]) * 0.1
        vedo_mesh = Sphere(pos=pos, r=radio).color(color)
        write(vedo_mesh, f"{export_dir}esfera_{i}.vtk")  # Formato nativo vedo

    # Cilindros con open3d (exportación individual)
    else:
        altura = 1.0 + abs(pos[2]) * 0.2
        cilindro = o3d.geometry.TriangleMesh.create_cylinder(
            radius=0.2, height=altura, resolution=20)
        cilindro.translate(pos)

        # Exportar con open3d
        o3d.io.write_triangle_mesh(
            f"{export_dir}cilindro_{i}.glb",
            cilindro
        )
        # Convertir a vedo
        vedo_mesh = Mesh([np.asarray(cilindro.vertices),
                        np.asarray(cilindro.triangles)]).color(color)

    # Aplicar rotación y añadir a la lista
    if pos[0] > 0:
        vedo_mesh.rotate(np.random.random()*360, axis=(0,1,0))

    vedo_objects.append(vedo_mesh)

# Exportar escena completa como GLTF (corregido)
escena_completa = merge(vedo_objects)  # Combinar todos los objetos
write(escena_completa, f"{export_dir}escena_completa.gltf")

# Visualización
show(escena_completa,
     axes=14,
     viewup='z',
     title="Escena Exportada",
     bg="white",
     bg2="lightblue",
     interactive=True)
```
